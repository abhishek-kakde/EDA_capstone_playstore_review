````python
# Write your code to make your dataset analysis ready.
# we need to remove 1474 rows of apps which has null rating as our analysis is based on customer rating
df.dropna(how='any', subset=['rating'], axis=0, inplace = True)
````

````python
# Filling Na values of Android version and current ver with mode of all as asumming it might be the same version as found in most apps.
df['android_ver'].fillna(df['android_ver'].mode()[0], inplace = True)
df['current_ver'].fillna(df['current_ver'].mode()[0], inplace = True)
````

````python
df.head(2)
````

````python
df.info()
````

````python
# reviews installs and price, size should be in integer or float
df['reviews'] = df['reviews'].astype(int)
````

````python
inst = []
for i in df['installs']:
  inst.append(i.replace(',','').replace('+','').strip())
df['installs'] = inst
df['installs'] = df['installs'].astype(int)
````

````python
#Removing $ sign
pricen = []
for p in df['price']:
  if p[0] == '$':
    pricen.append(p[1:])
  else:
    pricen.append(p)
df['price'] = pricen
df['price'] = df['price'].astype(float)
````

````python
df['rating'] = pd.to_numeric(df['rating'])
````

````python
# convert 'size' object into Float
#Base size is in Mb

l = [i for i in df['size']]

def size_float(size):
  clean_data=[]
  for i in size:
    i = str(i)
    if 'M' in i:
      i = i.replace('M' , '')
      i=float(i)
    elif 'k' in i:
      i= i.replace('k', '')
      i=float(i)
      i= i/1000
    elif 'Varies with device' in i:
      i=float(0)
    clean_data.append(i)
  return clean_data
df['size'] = size_float(l)
df['size']= df['size'].astype(float)
````

Outliers are data points that fall significantly outside the range of typical values. Box plots can help in identifying these extreme values that may have an impact on the overall analysis or interpretation of the data.
