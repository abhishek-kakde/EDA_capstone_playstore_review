````python
#Removing $ sign
pricen = []
for p in df['price']:
  if p[0] == '$':
    pricen.append(p[1:])
  else:
    pricen.append(p)
df['price'] = pricen
df['price'] = df['price'].astype(float)
````

````python
df['rating'] = pd.to_numeric(df['rating'])
````

````python
# convert 'size' object into Float
#Base size is in Mb

l = [i for i in df['size']]

def size_float(size):
  clean_data=[]
  for i in size:
    i = str(i)
    if 'M' in i:
      i = i.replace('M' , '')
      i=float(i)
    elif 'k' in i:
      i= i.replace('k', '')
      i=float(i)
      i= i/1000
    elif 'Varies with device' in i:
      i=float(0)
    clean_data.append(i)
  return clean_data
df['size'] = size_float(l)
df['size']= df['size'].astype(float)
````

* App Categories and Popularity:

* Sentiment Analysis:

Sentiment analysis on user reviews revealed that the majority of sentiments are positive. Users appreciate app features, usability, and functionality. Negative sentiments often revolve around performance issues, bugs, or specific feature requests.

* Impact of Size and Price:

Larger apps tend to have higher ratings, suggesting that users may associate larger file sizes with more extensive and feature-rich applications. Moreover, the majority of apps in the Play Store are free, with paid apps receiving generally positive reviews.
